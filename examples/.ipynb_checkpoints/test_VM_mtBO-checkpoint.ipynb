```python
import sys
sys.path.append('../')

import numpy as np
import matplotlib.pyplot as plt
import torch
from eBO import mtBO, virtual_machineIO, objective
from eBO.construct_machineIO import construct_machineIO
from eBO.utils import capture_warnings
```

```python
n_init = 5
n_eval = 5
beta = 4
```

```python
n_control = 4
n_task = 2

control_CSETs = [f'PSC{i}:I_CSET' for i in range(n_control)]
control_RDs = [f'PSC{i}:I_RD' for i in range(n_control)]
control_tols = [0.1]*len(control_CSETs)
control_min  = [0.0]*len(control_CSETs)
control_max  = [1.0]*len(control_CSETs)
monitor_PVs  = [f'BPM{i}:XPOS_RD' for i in range(n_task)]
```

```python
vm = virtual_machineIO.VM( 
    control_CSETs = control_CSETs,
    control_RDs= control_RDs,
    monitor_PVs= monitor_PVs,
    control_min= control_min,
    control_max= control_max,
    monitor_min= [-4]*n_task,
    monitor_max= [ 4]*n_task,
)

machineIO = construct_machineIO(
    virtual_machineIO=vm, 
    set_manually=False)
```

```python
mtIndexer = objective.MultiTaskIndexing(task_names=monitor_PVs)
input_names = [monitor_PVs[:1], monitor_PVs[:2]]
output_names = ['ObjPV1:Goal=1','objPV1,2:same']


iy1 = mtIndexer(input_names[0])
def obj1(y):
    return 1-(y[...,0] - 1)**2

iy2 = mtIndexer(input_names[1])
def obj2(y):
    return 1-(y[...,0] - y[...,1])**2

multi_task_objective_fn = objective.CompositeObjective(objfuncs=[obj1,obj2],
                                                       y_index=[iy1,iy2], 
                                                       weights=[0.5,0.5],
                                                       input_names = input_names,
                                                       output_names = output_names)
```

```python
with capture_warnings():
    mtbo = mtBO.MultiTaskBO(
        machineIO,
        control_CSETs = control_CSETs,
        control_RDs   = control_RDs,
        control_tols  = control_tols,
        control_min   = control_min,
        control_max   = control_max,
        monitor_PVs   = monitor_PVs,
        multi_task_objective_fn = multi_task_objective_fn,
        local_optimization = False,
        local_bound_size   = None,
        MultiTaskIndexer = mtIndexer
    )
```

```python
with capture_warnings():
    mtbo.init(n_init)
#     mtbo.plot_objective_RDs()
#     mtbo.plot_composite_objective()
# control_values = msbo.machineIO.history['set_order'][control_CSETs].iloc [:1].values
```
